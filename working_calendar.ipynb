```python
import json
import jsonfix
js = {}
fname = "ramesh"
with open(fname+".json", encoding="utf8") as f:
    js = json.loads(f.read())
js = jsonfix.fix_json(js)
x = {}
for i in js['lifeLog'].values():
    for j in i['mobile'].values():
        if j and 'calendar' in j:
            x[j['time']['timeStamp']] = j

```

```python
from datetime import datetime
working = []
for day, segments in js['segment'].items():
    print(day)
    # sorted segment numbers
    indexes = sorted(list(segments.keys()),key=lambda x:int(x))
    print(indexes)
    print(segments)
    for segment in range(len(indexes)):
        val = segments[indexes[segment]]
        if any(val['dailyActivitySet'][j]['KEY_DAILYACTIVITY'] == "working" for j in range(len(val['dailyActivitySet']))):
            if int(segment)+1 < len(indexes):
                working.append( (datetime.strptime(day+" "+val['labeledSegment'][0]['timeStamp'], "%m%d_%Y %H:%M"),\
                            datetime.strptime(day+" "+segments[str(int(indexes[segment+1]))]['labeledSegment'][0]['timeStamp'], "%m%d_%Y %H:%M")))
            else:
                working.append( (datetime.strptime(day+" "+val['labeledSegment'][0]['timeStamp'], "%m%d_%Y %H:%M"),\
                            datetime.strptime(day+" 23:59", "%m%d_%Y %H:%M")))

```

```python
working
```

```python
final_cal = {}
for i in x:
    dt = datetime.fromtimestamp(i/1000)
    if any([(j < dt < k) for j,k in working]):
        final_cal[i] = x[i]
```

```python
import os
os.makedirs("in", exist_ok=True)

with open("in/"+fname+"_working","w+") as f:
    f.write(json.dumps(final_cal))
with open("in/"+fname,"w+") as f:
    f.write(json.dumps(x))
```
